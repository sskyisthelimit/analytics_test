```python
dataset_path = "../assets/initial_dataset.csv"
```

```python
#adjust to unix style (it wouldn't solve the main problem)
with open(dataset_path, 'r', encoding='utf-8') as file:
    content = file.read()

content = content.replace('\r\n', '\n')

with open(dataset_path, 'w', encoding='utf-8') as file:
    file.write(content)
```
